```python
import numpy as np
import matplotlib.pyplot as plt

def x0(t):
    if 0 < t < 1:
        return 1
    elif 1 <= t < 2:
        return -t + 2
    else:
        return 0
     
t = np.linspace(-5, 5, 400)

x0 = [x0(t_val) for t_val in t]

plt.plot(t, x0, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Gráfica 0; x(t+1)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def x0(t):
    if 0 < t < 1:
        return 1
    elif 1 <= t < 2:
        return -t + 2
    else:
        return 0

def x1(t):
    if 0 < (t + 1) < 1:
        return 1
    elif 1 <= (t + 1) < 2:
        return -(t + 1) + 2
    else:
        return 0
    
def x2(t):
    if 0 < (-t + 1) < 1:
        return 1
    elif 1 <= (-t + 1) < 2:
        return -(-t + 1) + 2
    else:
        return 0
    
def x3(t):
    if 0 < (3/2 * t) < 1:
        return 1
    elif 1 <= (3/2 * t) < 2:
        return -(3/2 * t) + 2
    else:
        return 0
    
def x4(t):
    if 0 < (3/2 * t + 1) < 1:
        return 1
    elif 1 <= (3/2 * t + 1) < 2:
        return -(3/2 * t + 1) + 2
    else:
        return 0
    
t = np.linspace(-5, 5, 400)

x0 = [x0(t_val) for t_val in t]
x1 = [x1(t_val) for t_val in t]
x2 = [x2(t_val) for t_val in t]
x3 = [x3(t_val) for t_val in t]
x4 = [x4(t_val) for t_val in t]

plt.plot(t, x0, label='x(t)')
plt.plot(t, x1, label='x(t + 1)')
plt.plot(t, x2, label='x(-t + 1)')
plt.plot(t, x3, label='x(3/2t)')
plt.plot(t, x4, label='x(3/2t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Gráfico de punto 4')
plt.grid(True)
plt.legend()
plt.show()
```
