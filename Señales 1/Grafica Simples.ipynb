```python
import numpy as np
import matplotlib.pyplot as plt

def x0(t):
    if -2 < t < 2:
        return 1

    else:
        return 0
     
t = np.linspace(-5, 5, 400)

x0 = [x0(t_val) for t_val in t]

plt.plot(t, x0, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Gráfica x(t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def x0(t):
    if -2 < 2 * t < 2:
        return 1

    else:
        return 0
     
t = np.linspace(-3, 3, 400)

x0 = [x0(t_val) for t_val in t]

plt.plot(t, x0, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Gráfica x(2t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def x0(t):
    if -2 < (2 * t) + 2 < 2:
        return 1

    else:
        return 0
     
t = np.linspace(-3, 3, 400)

x0 = [x0(t_val) for t_val in t]

plt.plot(t, x0, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Gráfica x(2t)')
plt.grid(True)
plt.legend()
plt.show()
```
