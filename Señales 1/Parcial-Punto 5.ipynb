```python
## Sebastian Andres Zabala Saenz
import numpy as np
import matplotlib.pyplot as plt

def x(t):
    if -1 < (t) < 0:
        return 0
    elif 0 < (t) < 1:
        return (t) 
     
t = np.linspace(-2, 2, 400)

x = [x(t_val) for t_val in t]

plt.plot(t, x, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('GrÃ¡fica punto 5; x(t)')
plt.grid(True)
plt.legend()
plt.show()
```
