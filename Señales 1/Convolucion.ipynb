# Convolucion

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.signal import convolve

#
def x(t):
    return np.exp(3 * t) * (t <= 0)

def h(t):
    return (t >= 4)

# 
t = np.linspace(-5, 10, 1000)

#
convolution = convolve(x(t),h(t), mode='full') * (t[1] - t[0])


plt.figure(figsize=(10,5))
plt.plot(t, x(t), label = '$x(t)$')
plt.plot(t, x(t), label = '$h(t)$')
plt.plot(t, convolution[:len(t)], label='$(x*h)(t)$')
plt.title('Convolution de $x(t)$ y $h(t)$')
plt.xlabel('$t$')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show
```
