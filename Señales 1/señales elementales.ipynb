## Función escalon unitario  

### En tiempo continuo

```python
import numpy as np
import matplotlib.pyplot as plt

def u(t):
    if t > 0:
        return 1
    elif t < 0:
        return 0
     
t = np.linspace(-5, 5, 400)

u = [u(t_val) for t_val in t]

plt.plot(t, u, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Función escacalon unitario - Tiempo continuo')
plt.grid(True)
plt.legend()
plt.show()
```

### En tiempo discreto

```python
import numpy as np
import matplotlib.pyplot as plt

def v(n):
    return np.where(n >= 0, 1, 0)

n = np.arange(-10, 10)

u = v(n)

plt.stem(n, u, use_line_collection=True)
plt.title('Función Escalón unitario en tiempo discreto')
plt.xlabel('Tiempo discreto (n)')
plt.ylabel('u[n]')
plt.grid(True)
plt.ylim(-0.25, 1.5)
plt.show()
```

## Función rectangular 

```python
import numpy as np
import matplotlib.pyplot as plt

def rect(n, a, b):
    return np.where((n >= a) & (n <= b), 2, 0)

a = -5
b = 5

n = np.arange(-10, 10)

rect = rect(n, a, b)

plt.stem(n, rect, use_line_collection=True)
plt.title('Función Rectangular')
plt.xlabel('t')
plt.ylabel('Rect[n]')
plt.grid(True)
plt.ylim(-0.25, 2.5) 
plt.show()
```

## Funcion signo

```python
import numpy as np
import matplotlib.pyplot as plt

def r(t):
    if t < 0:
        return 0
    elif t > 0:
        return t
     
t = np.linspace(-1, 5, 400)

r = [r(t_val) for t_val in t]

plt.plot(t, r, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Función rampa')
plt.grid(True)
plt.legend()
plt.show()
```
