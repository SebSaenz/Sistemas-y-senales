## Grafica de series de Fourier

```python
import numpy as np
import matplotlib.pyplot as plt

# funcion
def F(t):
    return (- np.cos(2 * t)+1)

t = np.linspace(-np.pi, np.pi, 1000)

# serie de Fourier
def serie_d_fourier(N):
    result = 0
    for n in range(1, N + 1):
        result += (4 / (n**2 * np.pi**2)) * (-1)**n * np.cos(n * t)
    return result

# terminos en la serie
N = 10

# Gráfica de la serie de Fourier
plt.plot(t, F(t), label='Función original')
plt.plot(t, serie_d_fourier(N), label='Serie de Fourier (N={})'.format(N))
plt.xlabel('t')
plt.ylabel('F(t)')
plt.legend()
plt.show()
```

## Coeficiente de Serie de Fourier

```python
import numpy as np
from scipy.integrate import quad

# funcion X(t)
def X(t):
    return (- np.cos(2 * t)+1)

# Periodo
T =  np.pi

# a0
def a0():
    integral, _ = quad(X, -np.pi, np.pi)
    return (1 / T) * integral

# an
def an(n):
    integrand = lambda t: X(t) * np.cos((2 * n * np.pi * t) / T)
    integral, _ = quad(integrand, -np.pi, np.pi)
    return (2 / T) * integral

# bn
def bn(n):
    integrand = lambda t: X(t) * np.sin((2 * n * np.pi * t) / T)
    integral, _ = quad(integrand, -np.pi, np.pi)
    return (2 / T) * integral

# Ejemplo de calculo de algunos coeficientes
N = 1  # terminos en la serie de Fourier

# Calculando los coeficientes
coef_a0 = a0()
coef_an = [an(n) for n in range(1, N + 1)]
coef_bn = [bn(n) for n in range(1, N + 1)]

# Mostrar resultados
print("Coeficiente a0:", coef_a0)
print("Coeficientes an:", coef_an)
print("Coeficientes bn:", coef_bn)
```
