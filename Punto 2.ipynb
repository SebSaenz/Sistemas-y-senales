```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 1000)

x1 = np.exp(1j * (2 * np.pi / 3) * t)

plt.figure(figsize=(10, 6))
plt.plot(t, x1.real, label='Parte Real')
plt.plot(t, x1.imag, label='Parte Imaginaria')
plt.xlabel('t')
plt.ylabel('X1(t)')
plt.title('Grfica punto 2-a')
plt.legend()
plt.grid(True)
plt.show()
```
