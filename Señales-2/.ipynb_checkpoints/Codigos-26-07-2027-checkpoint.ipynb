# Ejemplo de codigo Modulacion en Amplitud 

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros

Ac = 1.0   # Amplitud de la portadora
fc = 50.0  # Frecuencia de la portadora en Hz
Am = 1.0   # Amplitud de la señal de mensaje
fm = 5.0   # Frecuencia de la señal de mensaje en Hz


t = np.linspace(0, 1, 1000)  # Vector de tiempo

# Señales
# portadora = Ac * np.cos(2 * np.pi * fc * t)
portadora = Ac * np.cos(2 * np.pi * fc * t)
mensaje = Am * np.cos(2 * np.pi * fm * t)


# Señal modulada en amplitud
modulada = (1 + mensaje) * portadora

# Envolvente de la señal AM
envolvente_superior = (1 + mensaje) * Ac
envolvente_inferior = -(1 + mensaje) * Ac

# Gráficas
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(t, mensaje, label='Señal de Mensaje')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal de Mensaje m(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, portadora, label='Señal Portadora')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Portadora c(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, modulada, label='Señal Modulada AM')
plt.plot(t, envolvente_superior, 'r-', label='Envolvente Superior')
plt.plot(t, envolvente_inferior, 'r-', label='Envolvente Inferior')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Modulada AM s(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# Señal de mensaje configurable 

```python
import numpy as np
import matplotlib.pyplot as plt

def am_modulation(Ac, Am, fc, fm, duration=1, sampling_rate=1000):
    t = np.linspace(0, duration, int(duration * sampling_rate))  # Vector de tiempo
    
    # Señales
    portadora = Ac * np.cos(2 * np.pi * fc * t)
    mensaje = Am * np.cos(2 * np.pi * fm * t)
    
    # Señal modulada en amplitud
    modulada = (1 + mensaje) * portadora
    
    # Gráficas
    plt.figure(figsize=(10, 8))

    # Envolvente de la señal AM
    envolvente_superior = (1 + mensaje) * Ac
    envolvente_inferior = -(1 + mensaje) * Ac

    plt.subplot(3, 1, 1)
    plt.plot(t, mensaje, label='Señal de Mensaje')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')
    plt.title('Señal de Mensaje m(t)')
    plt.grid(True)
    plt.legend()
    
    plt.subplot(3, 1, 2)
    plt.plot(t, portadora, label='Señal Portadora')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')
    plt.title('Señal Portadora c(t)')
    plt.grid(True)
    plt.legend()
    
    plt.subplot(3, 1, 3)
    plt.plot(t, modulada, label='Señal Modulada AM')
    plt.plot(t, envolvente_superior, 'r-', label='Envolvente Superior')
    plt.plot(t, envolvente_inferior, 'r-', label='Envolvente Inferior')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')
    plt.title('Señal Modulada AM s(t)')
    plt.grid(True)
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Solicitar parámetros al usuario

Ac = float(input("Introduce la amplitud de la portadora (Ac): "))
fc = float(input("Introduce la frecuencia de la portadora (fc) en Hz: "))
Am = float(input("Introduce la amplitud de la señal de mensaje (Am): "))
fm = float(input("Introduce la frecuencia de la señal de mensaje (fm) en Hz: "))



# Generar y visualizar la señal modulada
am_modulation(Ac, Am, fc, fm)
```
