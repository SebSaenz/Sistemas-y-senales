```python
# Importar bibliotecas necesarias para manipulación de imágenes, cálculos y GUI
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
from tkinter import filedialog, Scale

# Función para cargar la imagen desde una ruta de archivo
def cargar_imagen(ruta):
    """
    Carga una imagen y la convierte a formato RGB para trabajar en color.
    :param ruta: Ubicación del archivo de imagen.
    :return: La imagen en formato RGB.
    """
    imagen = Image.open(ruta).convert('RGB')  # Convertir a RGB para obtener 3 canales
    return imagen

# Función para mostrar la imagen en una ventana utilizando matplotlib
def mostrar_imagen(imagen):
    """
    Muestra la imagen usando matplotlib para verla sin ejes y en colores.
    :param imagen: Imagen a mostrar.
    """
    plt.imshow(imagen)
    plt.axis('off')  # Quitar los ejes para que se vea mejor
    plt.show()

# Función principal que aplica el ajuste RGB en el dominio de frecuencia
def aplicar_ajuste_frecuencia(imagen, r_factor, g_factor, b_factor):
    """
    Toma una imagen y ajusta los valores RGB utilizando la Transformada de Fourier (FFT).
    Básicamente, modifica todos los componentes de frecuencia de cada canal.
    
    :param imagen: Imagen original en formato RGB.
    :param r_factor: Factor de ajuste para el canal rojo.
    :param g_factor: Factor de ajuste para el canal verde.
    :param b_factor: Factor de ajuste para el canal azul.
    :return: Imagen con los canales ajustados en el dominio de frecuencia.
    """
    img_array = np.array(imagen)  # Convertimos la imagen a un array de numpy para manipular los píxeles

    # Creamos un array vacío para la imagen filtrada
    img_filtrada = np.zeros_like(img_array, dtype=np.float64)

    # Iteramos sobre cada canal (Rojo, Verde, Azul) y aplicamos FFT
    for c, factor in zip(range(3), [r_factor, g_factor, b_factor]):
        # Aplicamos FFT para pasar al dominio de frecuencia
        canal_fft = np.fft.fftshift(np.fft.fft2(img_array[..., c]))

        # Ajustamos la intensidad de todas las frecuencias multiplicándolas por el factor RGB correspondiente
        canal_fft_ajustado = canal_fft * factor

        # Transformamos de vuelta al dominio espacial con IFFT para obtener la imagen filtrada
        img_filtrada[..., c] = np.abs(np.fft.ifft2(np.fft.ifftshift(canal_fft_ajustado)))

    # Aseguramos que los valores estén en el rango [0, 255] para una imagen y convertimos a uint8
    img_filtrada = np.clip(img_filtrada, 0, 255).astype(np.uint8)
    return Image.fromarray(img_filtrada)

# Función para cargar imagen desde el GUI y aplicar el ajuste RGB en frecuencia
def cargar_y_aplicar_ajuste_gui():
    """
    Abre un diálogo para seleccionar la imagen desde el sistema de archivos
    y luego aplica el ajuste de frecuencia en RGB.
    """
    ruta = filedialog.askopenfilename()  # Abre el diálogo para seleccionar archivo
    if ruta:
        global imagen_original  # Guardamos la imagen original en una variable global
        imagen_original = cargar_imagen(ruta)  # Cargar la imagen
        aplicar_ajuste_rgb_gui()  # Aplica el ajuste RGB con la GUI

# Función que ajusta los valores RGB en frecuencia usando los sliders de la GUI
def aplicar_ajuste_rgb_gui():
    """
    Toma los valores de los sliders para cada canal RGB y aplica el ajuste de frecuencia.
    """
    if imagen_original:  # Si ya cargamos una imagen
        r_factor = r_slider.get()  # Obtener valor del slider para canal rojo
        g_factor = g_slider.get()  # Obtener valor del slider para canal verde
        b_factor = b_slider.get()  # Obtener valor del slider para canal azul
        # Aplicamos la transformación en frecuencia y mostramos la imagen
        imagen_ajustada = aplicar_ajuste_frecuencia(imagen_original, r_factor, g_factor, b_factor)
        mostrar_imagen(imagen_ajustada)

# Aquí comienza la interfaz gráfica con Tkinter
root = tk.Tk()
root.title("Editor de Imágenes con Ajuste en Frecuencia")

# Botón para cargar imagen
boton_cargar = tk.Button(root, text="Cargar Imagen", command=cargar_y_aplicar_ajuste_gui)
boton_cargar.pack()

# Sliders para ajustar los valores de RGB
r_slider = Scale(root, from_=0, to=2, resolution=0.1, orient=tk.HORIZONTAL, label="Factor Rojo")
r_slider.set(1)  # Valor por defecto
r_slider.pack()

g_slider = Scale(root, from_=0, to=2, resolution=0.1, orient=tk.HORIZONTAL, label="Factor Verde")
g_slider.set(1)  # Valor por defecto
g_slider.pack()

b_slider = Scale(root, from_=0, to=2, resolution=0.1, orient=tk.HORIZONTAL, label="Factor Azul")
b_slider.set(1)  # Valor por defecto
b_slider.pack()

# Botón para aplicar el ajuste RGB
boton_aplicar_ajuste = tk.Button(root, text="Aplicar Ajuste en Frecuencia", command=aplicar_ajuste_rgb_gui)
boton_aplicar_ajuste.pack()

# Inicializamos la variable de imagen como None
imagen_original = None

# Iniciar el bucle principal de la interfaz gráfica
root.mainloop()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
from tkinter import filedialog, Scale

# Función para cargar la imagen
def cargar_imagen(ruta):
    """
    Carga una imagen desde la ruta especificada y la convierte al formato RGB.
    
    Parámetros:
        ruta (str): Ruta del archivo de imagen que se desea cargar.
        
    Retorno:
        Image: Imagen en formato RGB.
    """
    imagen = Image.open(ruta).convert('RGB')  # Abrir y convertir la imagen a RGB
    return imagen

# Función para mostrar la imagen utilizando matplotlib
def mostrar_imagen(imagen):
    """
    Muestra la imagen usando matplotlib sin los ejes.
    
    Parámetros:
        imagen (Image): Imagen a visualizar.
        
    Retorno:
        Ninguno (solo muestra la imagen).
    """
    plt.imshow(imagen)
    plt.axis('off')  # Desactiva los ejes de la imagen
    plt.show()

# Función para aplicar el ajuste de frecuencia en los canales RGB de la imagen
def aplicar_ajuste_frecuencia(imagen, r_factor, g_factor, b_factor):
    """
    Aplica un ajuste de frecuencia a cada uno de los canales de color (rojo, verde y azul) de la imagen
    utilizando la Transformada Rápida de Fourier (FFT).
    
    Parámetros:
        imagen (Image): Imagen en formato RGB a la que se le aplicarán los ajustes.
        r_factor (float): Factor de ajuste para el canal rojo.
        g_factor (float): Factor de ajuste para el canal verde.
        b_factor (float): Factor de ajuste para el canal azul.
        
    Retorno:
        Image: Imagen ajustada en el dominio de frecuencia.
    """
    img_array = np.array(imagen)  # Convertir la imagen a un array de numpy

    # Inicializar la imagen filtrada que almacenará los resultados
    img_filtrada = np.zeros_like(img_array, dtype=np.float64)
    
    for c, factor in zip(range(3), [r_factor, g_factor, b_factor]):
        # Descomponer en frecuencia utilizando FFT
        canal_fft = np.fft.fftshift(np.fft.fft2(img_array[..., c]))

        # Aplicar el ajuste de frecuencia al canal
        canal_fft_ajustado = canal_fft * factor

        # Volver al dominio espacial utilizando la inversa de FFT
        img_filtrada[..., c] = np.abs(np.fft.ifft2(np.fft.ifftshift(canal_fft_ajustado)))

    # Escalar la imagen para asegurar que los valores no excedan 255 y convertirla a uint8
    img_filtrada = np.clip(img_filtrada, 0, 255).astype(np.uint8)

    # Asegurar que la imagen no tenga valores fuera del rango permitido (0-255)
    img_filtrada = np.clip(img_filtrada, 0, 255)

    return Image.fromarray(img_filtrada)

# Función para cargar la imagen y aplicar el ajuste RGB desde la interfaz gráfica
def cargar_y_aplicar_ajuste_gui():
    """
    Permite al usuario seleccionar una imagen mediante un diálogo de archivo
    y aplicar los ajustes de frecuencia a la imagen cargada.
    
    Retorno:
        Ninguno (realiza la acción dentro de la interfaz).
    """
    ruta = filedialog.askopenfilename()  # Abrir diálogo para seleccionar un archivo
    if ruta:
        global imagen_original
        imagen_original = cargar_imagen(ruta)  # Cargar la imagen seleccionada
        aplicar_ajuste_rgb_gui()  # Aplicar los ajustes de frecuencia con los deslizadores

# Función para aplicar el ajuste RGB utilizando los valores de los deslizadores
def aplicar_ajuste_rgb_gui():
    """
    Aplica los ajustes RGB de la imagen original basándose en los valores de los deslizadores
    de la interfaz gráfica y muestra la imagen resultante.
    
    Retorno:
        Ninguno (muestra la imagen ajustada).
    """
    if imagen_original:
        # Obtener los valores de los deslizadores para los factores de ajuste
        r_factor = r_slider.get()
        g_factor = g_slider.get()
        b_factor = b_slider.get()

        # Aplicar el ajuste en frecuencia con los factores seleccionados
        imagen_ajustada = aplicar_ajuste_frecuencia(imagen_original, r_factor, g_factor, b_factor)

        # Mostrar la imagen ajustada
        mostrar_imagen(imagen_ajustada)

# Interfaz gráfica usando Tkinter
root = tk.Tk()
root.title("Editor de Imágenes con Ajuste en Frecuencia")

# Botón para cargar la imagen desde el sistema de archivos
boton_cargar = tk.Button(root, text="Cargar Imagen", command=cargar_y_aplicar_ajuste_gui)
boton_cargar.pack()

# Deslizadores para ajustar los factores RGB de la imagen
r_slider = Scale(root, from_=0, to=2, resolution=0.1, orient=tk.HORIZONTAL, label="Factor Rojo")
r_slider.set(1)  # Valor inicial del deslizante para el canal rojo
r_slider.pack()

g_slider = Scale(root, from_=0, to=2, resolution=0.1, orient=tk.HORIZONTAL, label="Factor Verde")
g_slider.set(1)  # Valor inicial del deslizante para el canal verde
g_slider.pack()

b_slider = Scale(root, from_=0, to=2, resolution=0.1, orient=tk.HORIZONTAL, label="Factor Azul")
b_slider.set(1)  # Valor inicial del deslizante para el canal azul
b_slider.pack()

# Botón para aplicar los ajustes de frecuencia a la imagen
boton_aplicar_ajuste = tk.Button(root, text="Aplicar Ajuste en Frecuencia", command=aplicar_ajuste_rgb_gui)
boton_aplicar_ajuste.pack()

# Variable global para almacenar la imagen cargada
imagen_original = None

# Iniciar la interfaz gráfica
root.mainloop()
```
