# Ejemplo de codigo Modulacion en Amplitud 

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros

Ac = 1.0   # Amplitud de la portadora
fc = 50.0  # Frecuencia de la portadora en Hz
Am = 1.0   # Amplitud de la seal de mensaje
fm = 5.0   # Frecuencia de la seal de mensaje en Hz


t = np.linspace(0, 1, 1000)  # Vector de tiempo

# Seales
# portadora = Ac * np.cos(2 * np.pi * fc * t)
portadora = Ac * np.cos(2 * np.pi * fc * t)
mensaje = Am * np.cos(2 * np.pi * fm * t)


# Seal modulada en amplitud
modulada = (1 + mensaje) * portadora

# Envolvente de la seal AM
envolvente_superior = (1 + mensaje) * Ac
envolvente_inferior = -(1 + mensaje) * Ac

# Grficas
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(t, mensaje, label='Seal de Mensaje')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal de Mensaje m(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, portadora, label='Seal Portadora')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal Portadora c(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, modulada, label='Seal Modulada AM')
plt.plot(t, envolvente_superior, 'r-', label='Envolvente Superior')
plt.plot(t, envolvente_inferior, 'r-', label='Envolvente Inferior')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal Modulada AM s(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
Ac = 1.0   # Amplitud de la portadora
fc = 50.0  # Frecuencia de la portadora en Hz
Am = 1.0   # Amplitud de la seal de mensaje
fm = 5.0   # Frecuencia de la seal de mensaje en Hz

t = np.linspace(0, 1, 1000)  # Vector de tiempo

# Seales
portadora = Ac * np.cos(2 * np.pi * fc * t)
mensaje = Am * np.cos(2 * np.pi * fm * t)

# Seal modulada en amplitud
modulada = (1 + mensaje) * portadora

# Envolvente de la seal AM
envolvente_superior = (1 + mensaje) * Ac
envolvente_inferior = -(1 + mensaje) * Ac

# Transformada de Fourier de la seal modulada
fft_modulada = np.fft.fft(modulada)
frecuencias = np.fft.fftfreq(len(modulada), t[1] - t[0])

# Graficar resultados
plt.figure(figsize=(10, 10))

# Seal de mensaje
plt.subplot(4, 1, 1)
plt.plot(t, mensaje, label='Seal de Mensaje')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal de Mensaje m(t)')
plt.grid(True)
plt.legend()

# Seal portadora
plt.subplot(4, 1, 2)
plt.plot(t, portadora, label='Seal Portadora')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal Portadora c(t)')
plt.grid(True)
plt.legend()

# Seal modulada
plt.subplot(4, 1, 3)
plt.plot(t, modulada, label='Seal Modulada AM')
plt.plot(t, envolvente_superior, 'r-', label='Envolvente Superior')
plt.plot(t, envolvente_inferior, 'r-', label='Envolvente Inferior')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal Modulada AM s(t)')
plt.grid(True)
plt.legend()

# Espectro de frecuencias de la seal modulada
plt.subplot(4, 1, 4)
plt.plot(frecuencias, np.abs(fft_modulada), label='Espectro de Frecuencia')
plt.xlim(-100, 100)  # Limitar el eje x para ver mejor las componentes principales
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.title('Espectro de Frecuencia de la Seal Modulada')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros de la seal
Fs = 5000  # Frecuencia de muestreo (Hz)
T = 1      # Duracin de la seal (s)
t = np.linspace(0, T, int(Fs * T), endpoint=False)  # Vector de tiempo

# Parmetros de la seal moduladora
Am = 1       # Amplitud de la seal de mensaje
fm = 5       # Frecuencia de la seal de mensaje (Hz)
message = Am * np.sin(2 * np.pi * fm * t)  # Seal de mensaje (moduladora)

# Parmetros de la seal portadora
Ac = 1       # Amplitud de la portadora
fc = 100     # Frecuencia de la portadora (Hz)
kf = 100     # ndice de modulacin en radianes por segundo por voltio

# Seal modulada por frecuencia (FM)
integral_of_message = np.cumsum(message) / Fs
fm_signal = Ac * np.cos(2 * np.pi * fc * t + 2 * np.pi * kf * integral_of_message)

# Visualizacin de las seales
plt.figure(figsize=(12, 8))

# Seal de mensaje
plt.subplot(3, 1, 1)
plt.plot(t, message)
plt.title('Seal de Mensaje (Moduladora)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

# Seal portadora (sin modulacin)
plt.subplot(3, 1, 2)
plt.plot(t, Ac * np.cos(2 * np.pi * fc * t))
plt.title('Seal Portadora (Sin Modulacin)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

# Seal modulada por frecuencia (FM)
plt.subplot(3, 1, 3)
plt.plot(t, fm_signal)
plt.title('Seal Modulada por Frecuencia (FM)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
Ac = 1.0   # Amplitud de la portadora
fc = 50.0  # Frecuencia de la portadora en Hz
Am = 1.0   # Amplitud de la seal de mensaje
fm = 5.0   # Frecuencia de la seal de mensaje en Hz

t = np.linspace(0, 1, 1000)  # Vector de tiempo

# Seal de mensaje
mensaje = Am * np.cos(2 * np.pi * fm * t)

# Seal modulada en frecuencia (FM) sin necesidad de sensibilidad de frecuencia
modulada_fm = Ac * np.cos(2 * np.pi * fc * t + (Am / fm) * np.sin(2 * np.pi * fm * t))

# Transformada de Fourier de la seal modulada FM
fft_modulada_fm = np.fft.fft(modulada_fm)
frecuencias = np.fft.fftfreq(len(modulada_fm), t[1] - t[0])

# Graficar resultados
plt.figure(figsize=(10, 10))

# Seal de mensaje
plt.subplot(4, 1, 1)
plt.plot(t, mensaje, label='Seal de Mensaje')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal de Mensaje m(t)')
plt.grid(True)
plt.legend()

# Seal portadora (para referencia)
plt.subplot(4, 1, 2)
plt.plot(t, Ac * np.cos(2 * np.pi * fc * t), label='Seal Portadora')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal Portadora c(t)')
plt.grid(True)
plt.legend()

# Seal modulada en frecuencia (FM)
plt.subplot(4, 1, 3)
plt.plot(t, modulada_fm, label='Seal Modulada FM')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal Modulada FM s(t)')
plt.grid(True)
plt.legend()

# Espectro de frecuencias de la seal modulada FM
plt.subplot(4, 1, 4)
plt.plot(frecuencias, np.abs(fft_modulada_fm), label='Espectro de Frecuencia')
plt.xlim(-100, 100)  # Limitar el eje x para ver mejor las componentes principales
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.title('Espectro de Frecuencia de la Seal Modulada FM')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros comunes
fs = 1000  # Frecuencia de muestreo en Hz
t = np.arange(0, 1, 1/fs)  # Vector de tiempo de 1 segundo
fc = 100  # Frecuencia de la portadora en Hz
Am = 1  # Amplitud de la seal moduladora
Ac = 1  # Amplitud de la seal portadora
fm = 5  # Frecuencia de la seal moduladora en Hz

# Seal moduladora
mensaje = Am * np.sin(2 * np.pi * fm * t)

# Modulacin AM (Amplitud Modulada)
modulacion_am = (1 + mensaje) * Ac * np.cos(2 * np.pi * fc * t)

# Modulacin FM (Frecuencia Modulada)
kf = 50  # ndice de modulacin en FM
modulacion_fm = Ac * np.cos(2 * np.pi * fc * t + kf * np.sin(2 * np.pi * fm * t))

# Grficos
plt.figure(figsize=(12, 8))

# Seal moduladora
plt.subplot(3, 1, 1)
plt.plot(t, mensaje)
plt.title('Seal Moduladora')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

# Seal AM
plt.subplot(3, 1, 2)
plt.plot(t, modulacion_am)
plt.title('Seal Modulada en AM')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

# Seal FM
plt.subplot(3, 1, 3)
plt.plot(t, modulacion_fm)
plt.title('Seal Modulada en FM')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```
