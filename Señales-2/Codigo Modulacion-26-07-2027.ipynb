# Ejemplo de codigo Modulacion en Amplitud 

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros

Ac = 1.0   # Amplitud de la portadora
fc = 50.0  # Frecuencia de la portadora en Hz
Am = 1.0   # Amplitud de la señal de mensaje
fm = 5.0   # Frecuencia de la señal de mensaje en Hz


t = np.linspace(0, 1, 1000)  # Vector de tiempo

# Señales
# portadora = Ac * np.cos(2 * np.pi * fc * t)
portadora = Ac * np.cos(2 * np.pi * fc * t)
mensaje = Am * np.cos(2 * np.pi * fm * t)


# Señal modulada en amplitud
modulada = (1 + mensaje) * portadora

# Envolvente de la señal AM
envolvente_superior = (1 + mensaje) * Ac
envolvente_inferior = -(1 + mensaje) * Ac

# Gráficas
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(t, mensaje, label='Señal de Mensaje')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal de Mensaje m(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, portadora, label='Señal Portadora')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Portadora c(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, modulada, label='Señal Modulada AM')
plt.plot(t, envolvente_superior, 'r-', label='Envolvente Superior')
plt.plot(t, envolvente_inferior, 'r-', label='Envolvente Inferior')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Modulada AM s(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
Ac = 1.0   # Amplitud de la portadora
fc = 50.0  # Frecuencia de la portadora en Hz
Am = 1.0   # Amplitud de la señal de mensaje
fm = 5.0   # Frecuencia de la señal de mensaje en Hz

t = np.linspace(0, 1, 1000)  # Vector de tiempo

# Señales
portadora = Ac * np.cos(2 * np.pi * fc * t)
mensaje = Am * np.cos(2 * np.pi * fm * t)

# Señal modulada en amplitud
modulada = (1 + mensaje) * portadora

# Envolvente de la señal AM
envolvente_superior = (1 + mensaje) * Ac
envolvente_inferior = -(1 + mensaje) * Ac

# Transformada de Fourier de la señal modulada
fft_modulada = np.fft.fft(modulada)
frecuencias = np.fft.fftfreq(len(modulada), t[1] - t[0])

# Graficar resultados
plt.figure(figsize=(10, 10))

# Señal de mensaje
plt.subplot(4, 1, 1)
plt.plot(t, mensaje, label='Señal de Mensaje')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal de Mensaje m(t)')
plt.grid(True)
plt.legend()

# Señal portadora
plt.subplot(4, 1, 2)
plt.plot(t, portadora, label='Señal Portadora')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Portadora c(t)')
plt.grid(True)
plt.legend()

# Señal modulada
plt.subplot(4, 1, 3)
plt.plot(t, modulada, label='Señal Modulada AM')
plt.plot(t, envolvente_superior, 'r-', label='Envolvente Superior')
plt.plot(t, envolvente_inferior, 'r-', label='Envolvente Inferior')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Modulada AM s(t)')
plt.grid(True)
plt.legend()

# Espectro de frecuencias de la señal modulada
plt.subplot(4, 1, 4)
plt.plot(frecuencias, np.abs(fft_modulada), label='Espectro de Frecuencia')
plt.xlim(-100, 100)  # Limitar el eje x para ver mejor las componentes principales
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.title('Espectro de Frecuencia de la Señal Modulada')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
Fs = 5000  # Frecuencia de muestreo (Hz)
T = 1      # Duración de la señal (s)
t = np.linspace(0, T, int(Fs * T), endpoint=False)  # Vector de tiempo

# Parámetros de la señal moduladora
Am = 1       # Amplitud de la señal de mensaje
fm = 5       # Frecuencia de la señal de mensaje (Hz)
message = Am * np.sin(2 * np.pi * fm * t)  # Señal de mensaje (moduladora)

# Parámetros de la señal portadora
Ac = 1       # Amplitud de la portadora
fc = 100     # Frecuencia de la portadora (Hz)
kf = 100     # Índice de modulación en radianes por segundo por voltio

# Señal modulada por frecuencia (FM)
integral_of_message = np.cumsum(message) / Fs
fm_signal = Ac * np.cos(2 * np.pi * fc * t + 2 * np.pi * kf * integral_of_message)

# Visualización de las señales
plt.figure(figsize=(12, 8))

# Señal de mensaje
plt.subplot(3, 1, 1)
plt.plot(t, message)
plt.title('Señal de Mensaje (Moduladora)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

# Señal portadora (sin modulación)
plt.subplot(3, 1, 2)
plt.plot(t, Ac * np.cos(2 * np.pi * fc * t))
plt.title('Señal Portadora (Sin Modulación)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

# Señal modulada por frecuencia (FM)
plt.subplot(3, 1, 3)
plt.plot(t, fm_signal)
plt.title('Señal Modulada por Frecuencia (FM)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Solicitar parámetros al usuario
Ac = float(input("Ingrese la amplitud de la portadora (Ac): "))
fc = float(input("Ingrese la frecuencia de la portadora en Hz (fc): "))
Am = float(input("Ingrese la amplitud de la señal de mensaje (Am): "))
fm = float(input("Ingrese la frecuencia de la señal de mensaje en Hz (fm): "))

# Vector de tiempo
t = np.linspace(0, 1, 1000)

# Señal de mensaje (moduladora)
mensaje = Am * np.cos(2 * np.pi * fm * t)

# Integración de la señal de mensaje
integral_of_message = np.cumsum(mensaje) * (t[1] - t[0])

# Señal modulada por frecuencia (FM)
fm_signal = Ac * np.cos(2 * np.pi * fc * t + 2 * np.pi * integral_of_message)

# Gráficas
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(t, mensaje, label='Señal de Mensaje')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal de Mensaje m(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
portadora = Ac * np.cos(2 * np.pi * fc * t)
plt.plot(t, portadora, label='Señal Portadora')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Portadora c(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, fm_signal, label='Señal Modulada FM')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Modulada por Frecuencia s(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
Ac = 1.0   # Amplitud de la portadora
fc = 50.0  # Frecuencia de la portadora en Hz
Am = 1.0   # Amplitud de la señal de mensaje
fm = 5.0   # Frecuencia de la señal de mensaje en Hz

t = np.linspace(0, 1, 1000)  # Vector de tiempo

# Señal de mensaje
mensaje = Am * np.cos(2 * np.pi * fm * t)

# Señal modulada en frecuencia (FM) sin necesidad de sensibilidad de frecuencia
modulada_fm = Ac * np.cos(2 * np.pi * fc * t + (Am / fm) * np.sin(2 * np.pi * fm * t))

# Transformada de Fourier de la señal modulada FM
fft_modulada_fm = np.fft.fft(modulada_fm)
frecuencias = np.fft.fftfreq(len(modulada_fm), t[1] - t[0])

# Graficar resultados
plt.figure(figsize=(10, 10))

# Señal de mensaje
plt.subplot(4, 1, 1)
plt.plot(t, mensaje, label='Señal de Mensaje')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal de Mensaje m(t)')
plt.grid(True)
plt.legend()

# Señal portadora (para referencia)
plt.subplot(4, 1, 2)
plt.plot(t, Ac * np.cos(2 * np.pi * fc * t), label='Señal Portadora')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Portadora c(t)')
plt.grid(True)
plt.legend()

# Señal modulada en frecuencia (FM)
plt.subplot(4, 1, 3)
plt.plot(t, modulada_fm, label='Señal Modulada FM')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Modulada FM s(t)')
plt.grid(True)
plt.legend()

# Espectro de frecuencias de la señal modulada FM
plt.subplot(4, 1, 4)
plt.plot(frecuencias, np.abs(fft_modulada_fm), label='Espectro de Frecuencia')
plt.xlim(-100, 100)  # Limitar el eje x para ver mejor las componentes principales
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.title('Espectro de Frecuencia de la Señal Modulada FM')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros comunes
fs = 1000  # Frecuencia de muestreo en Hz
t = np.arange(0, 1, 1/fs)  # Vector de tiempo de 1 segundo
fc = 100  # Frecuencia de la portadora en Hz
Am = 1  # Amplitud de la señal moduladora
Ac = 1  # Amplitud de la señal portadora
fm = 5  # Frecuencia de la señal moduladora en Hz

# Señal moduladora
mensaje = Am * np.sin(2 * np.pi * fm * t)

# Modulación AM (Amplitud Modulada)
modulacion_am = (1 + mensaje) * Ac * np.cos(2 * np.pi * fc * t)

# Modulación FM (Frecuencia Modulada)
kf = 50  # Índice de modulación en FM
modulacion_fm = Ac * np.cos(2 * np.pi * fc * t + kf * np.sin(2 * np.pi * fm * t))

# Gráficos
plt.figure(figsize=(12, 8))

# Señal moduladora
plt.subplot(3, 1, 1)
plt.plot(t, mensaje)
plt.title('Señal Moduladora')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

# Señal AM
plt.subplot(3, 1, 2)
plt.plot(t, modulacion_am)
plt.title('Señal Modulada en AM')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

# Señal FM
plt.subplot(3, 1, 3)
plt.plot(t, modulacion_fm)
plt.title('Señal Modulada en FM')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```
