```python
import numpy as np
import matplotlib.pyplot as plt

# Solicitar parámetros al usuario
Ac = float(input("Ingrese la amplitud de la portadora (Ac): "))
fc = float(input("Ingrese la frecuencia de la portadora en Hz (fc): "))
Am = float(input("Ingrese la amplitud de la señal de mensaje (Am): "))
fm = float(input("Ingrese la frecuencia de la señal de mensaje en Hz (fm): "))

# Vector de tiempo
t = np.linspace(0, 1, 1000)

# Señales
portadora = Ac * np.cos(2 * np.pi * fc * t)
mensaje = Am * np.cos(2 * np.pi * fm * t)

# Señal modulada en amplitud
modulada = (1 + mensaje) * portadora

# Envolvente de la señal AM
envolvente_superior = (1 + mensaje) * Ac
envolvente_inferior = -(1 + mensaje) * Ac

# Transformada de Fourier de la señal modulada
fft_modulada = np.fft.fft(modulada)
frecuencias = np.fft.fftfreq(len(modulada), t[1] - t[0])

# Graficar resultados
plt.figure(figsize=(10, 10))

# Señal de mensaje
plt.subplot(4, 1, 1)
plt.plot(t, mensaje, label='Señal de Mensaje')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal de Mensaje m(t)')
plt.grid(True)
plt.legend()

# Señal portadora
plt.subplot(4, 1, 2)
plt.plot(t, portadora, label='Señal Portadora')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Portadora c(t)')
plt.grid(True)
plt.legend()

# Señal modulada con envolventes
plt.subplot(4, 1, 3)
plt.plot(t, modulada, label='Señal Modulada AM')
plt.plot(t, envolvente_superior, 'r--', label='Envolvente Superior')
plt.plot(t, envolvente_inferior, 'r--', label='Envolvente Inferior')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Modulada AM s(t)')
plt.grid(True)
plt.legend()

# Espectro de frecuencias de la señal modulada
plt.subplot(4, 1, 4)
plt.plot(frecuencias, np.abs(fft_modulada), label='Espectro de Frecuencia')
plt.xlim(-2 * fc, 2 * fc)  # Ajustar el límite del eje X para mostrar ambas mitades
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.title('Espectro de Frecuencia de la Señal Modulada')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
