```python
import numpy as np
import matplotlib.pyplot as plt

def Punto1():
   
    fig, ax = plt.subplots()
    
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)
    
    ax.set_xlabel('t')
    ax.set_ylabel('f(X)')
    
    plt.grid(True)
    plt.title('Taller P1-a')
    
    t = np.linspace(-10, 10, 400)
    y_vals = np.sin(3 * t) + 2 * np.cos(4 * t)
    ax.plot(t, y_vals, label='f(x) = sin(3x) + 2cos(4x)', color='blue')
    ax.legend()
    
    plt.show()
Punto1()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def Punto1():
   
    fig, ax = plt.subplots()
    
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)
    
    ax.set_xlabel('t')
    ax.set_ylabel('f(X)')
    
    plt.grid(True)
    plt.title('Taller P1- A + B?')
    
    t = np.linspace(-10, 10, 400)
    y_vals = np.sin(3 * t) + 2 * np.cos(4 * t) + np.sin(3 * t) + np.cos(np.pi* t)
    ax.plot(t, y_vals, label='f(x) = sin(3x) + 2cos(4x)', color='blue')
    ax.legend()
    
    plt.show()
Punto1()
```
