```python
import numpy as np
import matplotlib.pyplot as plt

def u(t):
    if t > 0:
        return 1
    elif t < 0:
        return 0
     
t = np.linspace(-5, 5, 400)

u = [u(t_val) for t_val in t]

plt.plot(t, u, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Funcion escacalon unitario - Tiempo continuo')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def x0(t):
    if t < 0:
        return 0
    elif t > 0:
        return t

     
t = np.linspace(-5, 5, 400)

x0 = [x0(t_val) for t_val in t]

plt.plot(t, x0, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Gráfica 0; x(t+1)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def Sa():
   
    fig, ax = plt.subplots()
    
    ax.set_xlim(-10, 10)
    ax.set_ylim(-2, 2)
    
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)
    
    ax.set_xlabel('t')
    ax.set_ylabel('Sa(x)')
    
    plt.grid(True)
    plt.title('Función muestreo Sa(x)')
    
    t = np.linspace(-10, 10, 400)
    y_vals = (np.sin(t))/(t)
    ax.plot(t, y_vals, label='Sa(x) = sen(x)/(x)', color='blue')
    ax.legend()
    
    plt.show()
Sa()
```
